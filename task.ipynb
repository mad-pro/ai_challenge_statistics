```python
import pandas as pd
import matplotlib.pyplot as plt

dict_faturamento = {
    'data_ref': ['2023-01-01', '2020-02-01', '2021-03-01', '2022-04-01', '2023-05-01', '2023-06-01', '2020-07-01', '2021-08-01', '2022-09-01', '2023-10-01', '2022-11-01', '2023-12-01'],
    'valor': [400000, 890000, 760000, 430000, 920000, 340000, 800000, 500000, 200000, 900000, 570000, 995000]
}
```

Com base nas aulas do módulo de Estatística, utilize os métodos da biblioteca Pandas para tratar o dicionário abaixo e:

```python
1 Trazer a média das vendas
```

```python
df_faturamento = pd.DataFrame(dict_faturamento);
df_faturamento

# calcular a média de vendas
media_vendas = round(df_faturamento['valor'].mean(), 2);

print(media_vendas)
```

Criar um gráfico de barras vertical mostrando o mês de referência e o valor

```python
# Verificar o tipo dos campos do DataFrame
print(df_faturamento.dtypes)

# converter o campo data_ref de object para datetime
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'])

# Create a vertical bar chart showing the reference month and value
df_faturamento.plot(kind='bar', x='data_ref', y='valor')
```

# Criar um gráfico de linhas mostrando o mês de referência e o valor.



```python


df_faturamento.plot.line(x='data_ref', y='valor')

plt.xlabel('Referencia mês x valor')
plt.ylabel('Valor')
plt.title('Referencia mês x valor')
plt.show()
```
